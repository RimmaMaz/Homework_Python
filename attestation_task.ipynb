```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
plt.plot(x, y)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax1, ax2 = ax.flatten()

ax1.plot(x, y)
ax2.hist(x, bins=4, edgecolor='red')

ax1.set_title('График № 1', fontsize=16, fontweight='bold', color='#112080')
ax2.set_title('График № 2', fontsize=16, fontweight='bold', color='#852080')

plt.subplots_adjust(wspace=0.5)

plt.show()
```
